```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc = pd.read_csv('b1abb33c4f.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig = plt.figure(figsize = (10, 10))

ax1 = fig.add_subplot(2, 3, 1) 
ax1 = sinasc['APGAR1'].value_counts().sort_index().plot.bar(xlabel = 'APGAR1')

ax2 = fig.add_subplot(2, 3, 2) 
ax2 = sinasc['APGAR5'].value_counts().sort_index().plot.bar(xlabel = 'APGAR5')

ax3 = fig.add_subplot(2, 3, 3) 
ax3 = sinasc['ESTCIVMAE'].value_counts().sort_index().plot.bar(xlabel = 'ESTCIVMAE')

ax4 = fig.add_subplot(2, 3, 4) 
ax4 = sinasc['SEXO'].value_counts().sort_index().plot.bar(xlabel = 'SEXO')

ax5 = fig.add_subplot(2, 3, 5) 
ax5 = sinasc['GESTACAO'].value_counts().sort_index().plot.bar(xlabel = 'GESTACAO')


plt.subplots_adjust(wspace = 0.3, hspace = 0.25)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(6, 4))

sns.barplot(x = 'CONSULTAS', y = 'APGAR5', data = sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(6, 10))
fig, axes = plt.subplots(2, 2)

plt.subplot(2, 2, 1)
sns.countplot(data=sinasc, x='SEXO')
plt.title('Frequência de Sexo')
plt.ylim(13000, 14000)


plt.subplot(2, 2, 2)
sns.countplot(data=sinasc, x='GESTACAO')
plt.title('Frequência de Gestação')
plt.xticks(rotation=90)


plt.subplot(2, 2, 3)
sns.countplot(data=sinasc, x='ESTCIVMAE')
plt.title('Frequência de Estado Civil da Mãe')
plt.xticks(rotation=45)


plt.subplot(2, 2, 4)
sns.countplot(data=sinasc, x='CONSULTAS')
plt.title('Frequência de Consultas')


plt.tight_layout()
plt.show()
```
