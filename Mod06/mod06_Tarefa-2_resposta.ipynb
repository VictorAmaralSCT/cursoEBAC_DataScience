```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc = pd.read_csv('b1abb33c4f.csv')
sinasc.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sns.boxplot(y = 'IDADEMAE', data = sinasc, saturation=1)
plt.title('Idade da Mãe')
plt.ylabel('Idade')
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all')
sns.boxplot(y = 'PESO', data = sinasc, saturation=1)
plt.title('PESO DO BEBÊ')
plt.ylabel('Peso (em gramas)')
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all')
sinasc['IDADEMAE'].plot.hist(bins = 20)
plt.title('Idade da Mãe')
plt.xlabel('Idade')
```

Utilizando o Seaborn

```python
plt.close('all')
sns.displot(data = sinasc 
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 20
            ,element = 'step'
            ,kde = True)
plt.title('Idade da Mãe')
plt.xlabel('Idade')
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('all')
sns.displot(data = sinasc 
            ,alpha = .25
            ,x = 'PESO'
            ,bins = 30
            ,element = 'step'
            ,kde = True)
plt.title('Peso do Bebê')
plt.xlabel('Peso (em gramas)')
```
